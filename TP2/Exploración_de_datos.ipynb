# Exploración y Descripción y de los datos



```python
from pathlib import Path

import numpy as np
import pandas as pd
import pyarrow as pa
```

```python
import pandas as pd

# URL del archivo en GitHub
url = "https://raw.githubusercontent.com/danielruizuleta/Proyecto-Grupo-Argos/main/dataset_alpha_betha.csv"


alpha_df = pd.read_csv(url)


print(alpha_df.head())

```

```python
alpha_df.info()
```

```python
alpha_df.head(15)
```

```python
alpha_df = alpha_df.replace("?", np.nan)

alpha_df.info()
```

```python

null_values = alpha_df.isnull().sum()


print("Cantidad de valores nulos por columna:")
print(null_values)
```

```python
# Identificar las variables numéricas
numerical_vars = alpha_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Identificar las variables categóricas (de tipo object)
categorical_vars = alpha_df.select_dtypes(include=['object']).columns.tolist()

# Identificar las variables booleanas
boolean_vars = alpha_df.select_dtypes(include=['bool']).columns.tolist()

# Imprimir los resultados
print("Variables numéricas:")
print(numerical_vars)
print("\nVariables categóricas:")
print(categorical_vars)
print("\nVariables booleanas:")
print(boolean_vars)
```

```python
# Primero, identificamos todas las columnas de tipo 'object'
object_columns = alpha_df.select_dtypes(include=['object']).columns

# Luego, convertimos esas columnas a 'category'
alpha_df[object_columns] = alpha_df[object_columns].astype('category')

print(alpha_df.dtypes)
```

```python

alpha_df['Demand'] = pd.to_numeric(alpha_df['Demand'], errors='coerce')


print(alpha_df.dtypes)
```

# Análisis de Variables del Dataset

# **Variables Categóricas**:

# **1. Partner**: Estado de relación
#    Descripción: Indica si el cliente tiene pareja.
#    Valores posibles:
#        - **Yes**: El cliente tiene pareja.
#        - **No**: El cliente no tiene pareja.

# **2. Dependents**: Presencia de dependientes
#    Descripción: Indica si el cliente tiene dependientes económicos
#    Valores posibles:
#        - **Yes**: El cliente tiene dependientes.
#        - **No**: El cliente no tiene dependientes.

# **3. Service1**: Servicio de teléfono
#    Descripción: Indica si el cliente tiene servicio telefónico.
#    Valores posibles:
#        - **Yes**: El cliente tiene servicio telefónico.
#        - **No**: El cliente no tiene servicio telefónico.

# **4. Service2**: Tipo de servicio telefónico
#    Descripción: Define si el cliente tiene servicio de teléfono y qué tipo.
#    Valores posibles:
#        - **"No phone service"**: El cliente no tiene servicio telefónico.
#        - **"No"**: El cliente no tiene un servicio especificado, pero no se menciona el tipo de servicio.

# **5. Security**: Seguridad de la cuenta
#    Descripción: Indica si el cliente tiene seguridad activada en su cuenta.
#    Valores posibles:
#        - **Yes**: El cliente tiene seguridad activada.
#        - **No**: El cliente no tiene seguridad activada.

# **6. OnlineBackup**: Respaldo en línea
#    Descripción: Indica si el cliente tiene respaldo en línea para proteger su información.
#    Valores posibles:
#        - **Yes**: El cliente tiene respaldo en línea.
#        - **No**: El cliente no tiene respaldo en línea.

# **7. DeviceProtection**: Protección de dispositivos
#    Descripción: Indica si el cliente tiene protección para sus dispositivos.
#    Valores posibles:
#        - **Yes**: El cliente tiene protección para sus dispositivos.
#        - **No**: El cliente no tiene protección para sus dispositivos.

# **8. TechSupport**: Soporte técnico
#    Descripción: Indica si el cliente tiene acceso a soporte técnico.
#    Valores posibles:
#        - **Yes**: El cliente tiene acceso a soporte técnico.
#        - **No**: El cliente no tiene soporte técnico.

# **9. Contract**: Tipo de contrato
#    Descripción: Especifica el tipo de contrato que el cliente tiene con la empresa.
#    Valores posibles:
#        - **Month-to-month**: Contrato mensual.
#        - **One year**: Contrato anual.

# **10. PaperlessBilling**: Facturación sin papel
#     Descripción: Indica si el cliente prefiere recibir sus facturas electrónicas (sin papel).
#     Valores posibles:
#         - **Yes**: El cliente recibe facturación sin papel.
#         - **No**: El cliente recibe facturación en papel.

# **11. PaymentMethod**: Método de pago
#     Descripción: Define el método de pago utilizado por el cliente.
#     Valores posibles:
#         - **"Electronic check"**: El cliente paga mediante cheque electrónico.
#         - **"Mailed check"**: El cliente paga mediante cheque por correo.
#         - **"Bank transfer (automatic)"**: El cliente paga mediante transferencia bancaria automática.

# **12. Class**: Clase del cliente
#     Descripción: Define a qué clase pertenece el cliente.
#     Valores posibles:
#         - **Alpha**: Clase Alpha.
