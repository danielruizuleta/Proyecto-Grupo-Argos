# Exploración y Descripción y de los datos



```python
from pathlib import Path

import numpy as np
import pandas as pd
import pyarrow as pa
```

```python


# Directorio de trabajo actual
print("Directorio de trabajo actual:", Path.cwd())

DATA_DIR = Path.cwd() / "data" / "01_raw"

DATA_DIR.mkdir(parents=True, exist_ok=True)

# Cargar el archivo CSV en un DataFrame
alpha_df = pd.read_csv('/content/dataset_alpha_betha.csv')

# Ruta completa donde se guardará el archivo
file_path = DATA_DIR / "dataset_alpha_betha.csv"


alpha_df.to_csv(file_path, index=False)


print(f"Archivo guardado en: {file_path}")
```

```python
alpha_df.info()
```

```python
alpha_df.head(15)
```

```python
alpha_df = alpha_df.replace("?", np.nan)

alpha_df.info()
```

```python

null_values = alpha_df.isnull().sum()


print("Cantidad de valores nulos por columna:")
print(null_values)
```

```python
# Identificar las variables numéricas
numerical_vars = alpha_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Identificar las variables categóricas (de tipo object)
categorical_vars = alpha_df.select_dtypes(include=['object']).columns.tolist()

# Identificar las variables booleanas
boolean_vars = alpha_df.select_dtypes(include=['bool']).columns.tolist()

# Imprimir los resultados
print("Variables numéricas:")
print(numerical_vars)
print("\nVariables categóricas:")
print(categorical_vars)
print("\nVariables booleanas:")
print(boolean_vars)
```

```python
# Guardar el DataFrame en formato Parquet
alpha_df.to_parquet(DATA_DIR / "dataset_alpha_betha.parquet", index=False, engine='pyarrow')

print(f"Archivo guardado en: {DATA_DIR / 'dataset_alpha_betha.parquet'}")
```
